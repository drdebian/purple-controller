```python
import matplotlib.pyplot as plt
plt.style.use('default')
from matplotlib.patches import Rectangle
```

```python
def recplot(ax, x, y, w, h, color, label, hatch):
    ax.add_patch(Rectangle((x, y), w, h,
                           edgecolor=color,
                           fill=False,
                           hatch=hatch,
                           lw=1))
    ax.annotate(label,  (x + w/2, y + h/2),
                rotation=0, bbox=dict(facecolor=color, boxstyle='round'))
```

```python
fig, ax = plt.subplots(figsize=(10,7))

# Plotten
ax.plot()
recplot(ax,   0,  0,   1,   1, 'white',       '', '')
recplot(ax,   0,  0,   .2,  1, 'pink', '1', '/')
recplot(ax,  .2,  0,   .2, .75, 'orange',  '2', '\\')
recplot(ax,  .4,  0,   .2,  .5, 'magenta', '3', '/')
recplot(ax,  .6,  0,   .2,  .25, 'lightgreen', '4', '\\')
recplot(ax,  .8,  0,   .2,  0, 'lightblue', '5', '/')
plt.xlabel('$\\frac{Ladestand}{E_{max}}$')
plt.ylabel('$0=P_{min}, 1=P_{max}$')
plt.title('Ladeleistungsbegrenzung SoC->kW')
```
